<h1>Academic Integrity</h1>
    <p><b>Academic misconduct includes</b></p>
    <ul>
        <li><b>Poor Academic Practice</b>: Actions that include, but are not limited to, poor academic writing skills (e.g., poor referencing or passing off someone else's idea as your own), or small errors made through carelessness or misunderstanding.</li>
        <li><b>Cheating</b>: Actions that attempt to get advantage by means that undermine values of integrity.</li>
        <li><b>Contract Cheating / Essay Mills</b>: A form of academic misconduct where a person uses an undeclared and/or unauthorised third party, online or directly, to assist them in dishonestly producing work for academic credit or progression, whether or not payment or other favour is involved.</li>
        <li><b>Cumulation eﬀect</b>: Where continued poor academic practice and repeated minor instances of academic misconduct are treated as a case of major academic misconduct.</li>
        <li><b>Fabrication/Falsification</b>: Making up data, experiments, or other significant information in proposing, conducting, or reporting research. This includes the fabrication or falsification of official University documents regarding credit and/or academic achievement.</li>
        <li><b>Impersonation</b>: Undertaking in whole or in part any work required as part of a programme in the place of an enrolled learner, without permission from the provider.</li>
        <li><b>Unethical Use of Generative Artificial Intelligence (GenAI)</b>: Generative Artificial Intelligence (GenAI) refers to a subset of artificial intelligence (AI) that uses algorithms and models trained on massive datasets to generate new text, audio, video, code, and more. Academic integrity is breached if students submit the products of GenAI as their own work without acknowledgement and without authorisation to use GenAI in fulﬁlling the task.</li>
        <li><b>Plagiarism</b>: Presenting work or ideas taken from other sources without proper acknowledgement, whether done deliberately, carelessly, or inadvertently.</li>
        <li>Other acts that dishonestly use information to gain academic credit.</li>
    </ul>

<h1>Plagiarism</h1>
    <p><b>Presenting work or ideas taken from other sources without proper acknowledgement, whether done deliberately, carelessly, or inadvertently.</b></p>
    <p>It is a violation of UCC Policy and there are strict and severe penalties.</p>
    <p>Types of Plagiarism include but are not limited to:</p>
    <ul>
        <li><b>Collusion</b>: A joint effort of work is presented by an individual without due recognition of the input of others. Collusion also applies to both parties when an individual student provides their work to another student and allows them to present it as their own.</li>
        <li><b>Self-plagiarism</b>: The use of one's own previous submitted/presented work in another context without appropriate citation.</li>
        <li><b>Verbatim plagiarism</b>: Word-for-word copy of work from another source without providing acknowledgement.</li>
    </ul>
    <p>You must read and comply with the <a href="https://www.ucc.ie/en/academicgov/policies/standards/academicintegrityforexaminationsandassessmentspolicy2025-2026/">Academic Integrity for Examinations and Assessments Policy 2025-2026</a>.</p>
    <p>The Policy applies to all work submitted, including software.</p>
    <p>You can expect that your work will be checked for evidence of plagiarism or collusion.</p>
    <p>In some circumstances it may be acceptable to reuse a small amount of work by others, but only if you provide explicit acknowledgement and justification.</p>
    <p>If in doubt ask your module lecturer prior to submission. Better safe than sorry!</p>
