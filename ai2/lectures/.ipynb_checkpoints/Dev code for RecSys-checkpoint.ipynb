```python
user_ids = ratings["user_id"].unique()
item_ids = ratings["item_id"].unique()
num_users = len(user_ids)
num_items = len(item_ids)
num_ratings = len(ratings)
mean = ratings["rating"].mean()
sparsity = 1 - num_ratings / (num_users * num_items)
num_users, num_items, num_ratings, mean, sparsity
```

```python
ratings_matrix = ratings.pivot(index="user_id", columns="item_id", values="rating").fillna(0)
```

```python
ratings_matrix
```

```python
def predict_rating_by_mean(user_id, item_id):
    # Ignores the user_id! Not personalised
    return ratings_matrix[ratings_matrix[item_id] != 0.0][item_id].mean()
```

```python
def sim(user_1_id, user_2_id):
    # Get the users' ratings for co-rated items
    user_1_ratings = []
    user_2_ratings = []
    for (r1, r2) in zip(ratings_matrix.loc[user_1_id], ratings_matrix.loc[user_2_id]):
        if r1 != 0.0 and r2 != 0.0:
            user_1_ratings.append(r1)
            user_2_ratings.append(r2)
    # Return Pearson correlation, or zero if Pearson cannot be calculated
    try:
        pearson = pearsonr(user_1_ratings, user_2_ratings)[0]
        return 0.0 if isnan(pearson) else pearson
    except:
        return 0.0
```

```python
def predict_rating_by_user_based_nn(user_id, item_id, k, ratings_matrix):
    # Get the user_ids sorted by similarity to the active user
    sorted_user_ids = np.argsort(
        [sim(user_id, u_id) for u_id in user_ids]).tolist()
    # To be a neighbour, you must have watched this movie 
    taboo_user_ids = ratings_matrix[ratings_matrix[item_id] == 0.0].index.tolist()
    # ...and you must be different from the active user
    taboo_user_ids.append(user_id)
    # Get the k neighbours
    neighbour_user_ids = [u_id for u_id in sorted_user_ids if u_id not in taboo_user_ids][:k]
    # Get the neighbours' ratings of this item
    neighbour_ratings = ratings_matrix.loc[neighbour_user_ids][item_id]
    # Return the mean of the neighours' ratings
    return neighbour_ratings.mean()
```

```python
def matrix_factorization(d, alpha, num_epochs, ratings_matrix):
    
    P = np.random.rand(num_users, d)
    Q = np.random.rand(num_items, d)
    
    for epoch in range(num_epochs):
        for i in range(num_users):
            for j in range(num_items):
                r = ratings_matrix.loc[i][j]
                if r != 0.0:
                    eij = mean + np.dot(P[i,:],Q[j,:]) - r
                    P[i] = P[i] - alpha * eij * Q[j]
                    Q[j] = Q[j] - alpha * eij * P[i]
                    
    return P, Q
```

```python
def predict_rating_by_mf(user_id, item_id, mean, P, Q):
    return mean + np.dot(P[user_id,:],Q[item_id,:])
```

```python
P, Q = matrix_factorization(d=10, alpha=0.001, num_epochs=15, ratings_matrix=ratings_matrix)
```

```python
predict_rating_by_mean(user_id=11, item_id=2), \
predict_rating_by_user_based_nn(user_id=11, item_id=12, k=50, ratings_matrix=ratings_matrix)
predict_rating_by_mf(user_id=11, item_id=2, mean=mean, P=P, Q=Q)
```

```python
# Shuffle the dataset
ratings = ratings.sample(frac=1, random_state=2)
ratings.reset_index(drop=True, inplace=True)
```

```python
# Split into train and test
train_size = int(0.8 * len(ratings))
train_ratings = ratings[:train_size]
test_ratings = ratings[train_size:]
```

```python
# Get the mean of the train ratings; convert into a ratings matrix; perform matrix factorization
train_mean = train_ratings["rating"].mean()
train_ratings_matrix = ratings.pivot(index="user_id", columns="item_id", values="rating").fillna(0)
train_P, train_Q = matrix_factorization(
    d=10, alpha=0.001, num_epochs=15, ratings_matrix=train_ratings_matrix)
```

```python
# Predict each test set rating and compute MAE for each recommender system
predictions = np.zeros(3)
maes = np.zeros(3)
for (u, i, r) in test_ratings.values:
    predictions = [predict_rating_by_mean(u, i), 
                   predict_rating_by_user_based_nn(u, i, k=50, ratings_matrix=train_ratings_matrix),
                   predict_rating_by_mf(u, i, mean=train_mean, P=train_P, Q=train_Q)]
    maes += np.abs(predictions - r)
print(maes / len(test_ratings))
```
